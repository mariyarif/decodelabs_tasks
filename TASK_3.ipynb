```python
# Import libraries
import pandas as pd

# Load dataset
df = pd.read_csv('/content/sample_data/matches.csv')
# Use cleaned dataset

# -------------------------------
# 1. BASIC STATISTICS
# -------------------------------

print("Basic Statistics:\n")
display(df.describe())

# -------------------------------
# 2. MATCHES PER SEASON
# -------------------------------

print("\nMatches per Season:")
season_counts = df['season'].value_counts().sort_index()
print(season_counts)

# -------------------------------
# 3. MOST WINNING TEAMS
# -------------------------------

print("\nTop Winning Teams:")
top_winners = df['winner'].value_counts().head(10)
print(top_winners)

# -------------------------------
# 4. MOST PLAYER OF MATCH AWARDS
# -------------------------------

print("\nTop Players (Player of Match):")
top_players = df['player_of_match'].value_counts().head(10)
print(top_players)

# -------------------------------
# 5. TOSS DECISION ANALYSIS
# -------------------------------

print("\nToss Decision Distribution:")
print(df['toss_decision'].value_counts())

# -------------------------------
# 6. SUPER OVER MATCHES
# -------------------------------

print("\nSuper Over Matches:")
print(df['super_over'].value_counts())

# -------------------------------
# 7. OUTLIER DETECTION
# -------------------------------

print("\nOutliers in Result Margin:")
print("Max Margin:", df['result_margin'].max())
print("Min Margin:", df['result_margin'].min())

# -------------------------------
# 8. VENUE ANALYSIS
# -------------------------------

print("\nTop Venues:")
print(df['venue'].value_counts().head(10))
```

```python
# STEP 1: Import libraries
import pandas as pd

# STEP 2: Load dataset
# change file name if needed
df = pd.read_csv('/content/sample_data/matches.csv')

# STEP 3: View first 5 rows
print("First 5 rows:")
display(df.head())

# STEP 4: Dataset shape
print("\nDataset Shape (rows, columns):")
print(df.shape)

# STEP 5: Column names
print("\nColumn Names:")
print(df.columns.tolist())

# STEP 6: Dataset info (data types + missing values)
print("\nDataset Info:")
df.info()

# STEP 7: Statistical summary
print("\nStatistical Summary:")
display(df.describe())

# STEP 8: Check missing values
print("\nMissing Values:")
print(df.isnull().sum())

# STEP 9: Unique values (for understanding categorical data)
print("\nUnique Values per Column:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} unique values")

# STEP 10: Sample random rows
print("\nRandom Sample:")
display(df.sample(5))
```

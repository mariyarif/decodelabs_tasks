```python
# -------------------------------
# IMPORT LIBRARIES
# -------------------------------
import pandas as pd
import matplotlib.pyplot as plt

# -------------------------------
# UPLOAD DATASET (run this first)
# -------------------------------
from google.colab import files
uploaded = files.upload()

# -------------------------------
# LOAD DATASET
# -------------------------------
df = pd.read_csv('matches.csv')

# (Optional) make a copy
df = df.copy()

# -------------------------------
# 1. MATCHES PER SEASON
# -------------------------------
plt.figure()
df['season'].value_counts().sort_index().plot(kind='bar')
plt.title("Matches Played per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 2. TOP 10 WINNING TEAMS
# -------------------------------
plt.figure()
df['winner'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Winning Teams")
plt.xlabel("Teams")
plt.ylabel("Wins")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 3. TOSS DECISION DISTRIBUTION
# -------------------------------
plt.figure()
df['toss_decision'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Toss Decision Distribution")
plt.ylabel("")
plt.show()

# -------------------------------
# 4. TOP PLAYERS (PLAYER OF MATCH)
# -------------------------------
plt.figure()
df['player_of_match'].value_counts().head(10).plot(kind='bar')
plt.title("Top Players of the Match Awards")
plt.xlabel("Players")
plt.ylabel("Awards Count")
plt.xticks(rotation=45)
plt.show()

# -------------------------------
# 5. RESULT MARGIN DISTRIBUTION
# -------------------------------
plt.figure()
df['result_margin'].plot(kind='hist', bins=20)
plt.title("Distribution of Match Result Margins")
plt.xlabel("Margin")
plt.ylabel("Frequency")
plt.show()
```

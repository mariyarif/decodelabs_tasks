```python
# Import libraries
import pandas as pd

# Load dataset
df = pd.read_csv('/content/sample_data/matches.csv')

# Make a copy (safe practice)
df_clean = df.copy()

# -------------------------------
# 1. HANDLE MISSING VALUES
# -------------------------------

# Check missing values
print("Missing Values Before Cleaning:\n", df_clean.isnull().sum())

# Fill missing city with 'Unknown'
df_clean['city'].fillna('Unknown', inplace=True)

# Fill missing player_of_match with 'Not Awarded'
df_clean['player_of_match'].fillna('Not Awarded', inplace=True)

# Fill missing winner with 'No Result'
df_clean['winner'].fillna('No Result', inplace=True)

# Fill numerical columns with median
df_clean['result_margin'].fillna(df_clean['result_margin'].median(), inplace=True)
df_clean['target_runs'].fillna(df_clean['target_runs'].median(), inplace=True)
df_clean['target_overs'].fillna(df_clean['target_overs'].median(), inplace=True)

# Drop 'method' column (too many missing values)
df_clean.drop(columns=['method'], inplace=True)

# -------------------------------
# 2. REMOVE DUPLICATES
# -------------------------------

before = df_clean.shape[0]
df_clean.drop_duplicates(inplace=True)
after = df_clean.shape[0]

print(f"\nDuplicates Removed: {before - after}")

# -------------------------------
# 3. FORMAT DATA TYPES
# -------------------------------

# Convert date column to datetime
df_clean['date'] = pd.to_datetime(df_clean['date'])

# Check final data types
print("\nData Types After Formatting:\n", df_clean.dtypes)

# -------------------------------
# 4. FINAL CHECK
# -------------------------------

print("\nMissing Values After Cleaning:\n", df_clean.isnull().sum())

print("\nFinal Dataset Shape:", df_clean.shape)

df_clean.head()
```
